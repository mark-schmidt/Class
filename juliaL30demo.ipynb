```julia
using PyCall
using PyPlot
plt = pyimport("matplotlib.pyplot")
SkManifold = pyimport("sklearn.manifold")
```

## t-SNE demo

```julia
SkDatasets = pyimport("sklearn.datasets")

data = SkDatasets.load_digits()
```

```julia
i = 10
plt.imshow(data["images"][i,:,:],cmap="gray")
target = data["target"][i]
plt.title("This is a $target")
```

```julia
X = data["data"]
y = data["target"];
```

```julia
tsne = SkManifold.TSNE()
Z = tsne.fit_transform(X);
```

```julia
plt.scatter(Z[:,1], Z[:,2], alpha=0.3);
plt.xlabel("z_1");
plt.ylabel("z_2");
```

```julia
plt.scatter(Z[:,1], Z[:,2], alpha=0.3, c=y);
plt.xlabel("z_1");
plt.ylabel("z_2");
```
